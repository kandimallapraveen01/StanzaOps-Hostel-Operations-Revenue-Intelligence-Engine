```python
import pandas as pd
df = pd.read_csv("Hostel_Management_Dataset_Filled.csv")
print(df.head())
print(df.dtypes)
```

```python
date_columns = ["Date_of_Birth", "Check_In_Date", "Payment_Date", "Joining_Date"]
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
print(df[date_columns].dtypes)
```

```python
print(df.columns.tolist())
```

```python
from datetime import datetime
today = pd.Timestamp.today()
df["Age"] = ((today - df["Date_of_Birth"]).dt.days // 365)
print(df[["Date_of_Birth", "Age"]].head())
```

```python
invalid_age = df[(df["Age"] < 0) | (df["Age"] > 100)]
print("Invalid Age Records:", len(invalid_age))
invalid_age
```

```python
invalid_deposit = df[df["Deposit_Amount"] < 0]
print("Invalid Deposit Amount Records:", len(invalid_deposit))
invalid_deposit
```

```python
invalid_paid = df[
    (df["Amount_Paid"] < 0) |
    (df["Amount_Paid"] > df["Deposit_Amount"])
]
print("Invalid Amount Paid Records:", len(invalid_paid))
invalid_paid
```
